```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data")
df.head(5)
```

```python
# Check data shape and missing values
print("Shape:", df.shape)
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
round(df.describe(), 2)
```

```python
# Box plots: Diabetic vs Non-diabetic comparison
fig, axes = plt.subplots(3, 3, figsize=(14, 10))
axes = axes.flatten()

features = df.columns[:-1]  # All columns except Outcome

for i, feature in enumerate(features):
    df.boxplot(column=feature, by='Outcome', ax=axes[i])
    axes[i].set_title(feature)
    axes[i].set_xlabel('Diabetes (0=No, 1=Yes)')
    axes[i].set_ylabel(feature)

plt.suptitle('Feature Comparison: Diabetic vs Non-Diabetic', fontsize=14, y=1.00)
plt.tight_layout()
plt.show()
```
